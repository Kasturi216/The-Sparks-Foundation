<font color='purple'><li><b>Importing Required Libraries</b></li></font>

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns

print('Libraries Imported Successfully!')
```

<font color='purple'><li><b>Loading the Dataset</b></li></font>

```python
loc='C:/Users/KASTURI/Downloads/Iris.csv'
data=pd.read_csv(loc)
print('Data Loaded Successfully!')
```

<font color='purple'><li><b>Exploring Data</b></li></font>

```python
data.head()
```

```python
data.shape
```

<b>There are 150 rows and 6 columns in the dataset.</b>

```python
data.info()
```

<b>Here, it is evident that, there are no null values present in the dataset.\
Also, We do not need the 'Id' column in our further analysis, so we would like to remove this column.</b>

```python
data.drop(['Id'], axis=1, inplace=True)
data
```

```python
data.describe(include='all')
```

<font color='purple'><li><b>Checking the relationships among attributes</b></li></font>

```python
sns.pairplot(data, hue='Species')
plt.show()
```

<b>We can conclude from the above plot that, Iris-Setosa is easily distinguishable from the rest, whereas Iris-versicolor\
and Iris-virginica are overlapping.</b>

<font color='purple'><li><b>Preparing Data</b></li></font>

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,4].values
```

<font color='purple'><li><b>Training the model</b></li></font>

```python
#Splitting Data

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=0)
print('The dataset has been split into 75% train data and 25% test data.')
```

```python
#Training the model

dtc=DecisionTreeClassifier(criterion='entropy', random_state=0)
dtc.fit(X_train, y_train)
print('Training Completed!')
```

```python
#Predicting the model

y_pred=dtc.predict(X_test)
y_pred
```

<font color='purple'><li><b>Model Evaluation</b></li></font>

<li><b>Confusion Matrix</b></li>

```python
confusion_matrix(y_test,y_pred)
```

<li><b>Classification Report</b></li>

```python
print(classification_report(y_test,y_pred))
```

<li><b>Accuracy Score</b></li>

```python
print('Accuracy Score: %0.2f'%accuracy_score(y_test,y_pred))
```

<b>We see that the Precision, Recall, F1-Score are above 90% also we have 97% accuracy score\
    which means the model is fitted well and works nicely.</b>

<font color='purple'><li><b>Visualizing the Decision Tree</b></li></font>

```python
fn=data.columns.values[0:4]
cn= data['Species'].unique()
print(fn)
print('\n')
print(cn)
```

```python
fig, axes=plt.subplots(nrows=1, ncols=1, figsize=(5,5), dpi=300)
tree.plot_tree(dtc, feature_names=fn, class_names=cn, filled=True)
plt.show()
```
